```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Heart Attack Dataset.csv")
```

```python
print(df.shape)
print(df.columns)
print(df.dtypes)
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df.describe(include='all')
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df.hist(figsize=(20,15))
plt.show()
```

```python
sns.countplot(x="Smoking",
hue="Heart_Attack_Risk",
data=df)
plt.show()
```

```python
plt.figure(figsize=(20,18))
sns.heatmap(df.select_dtypes(include='number').corr(),
            annot=True,
            cmap='coolwarm')
plt.show()
```

```python
sns.boxplot(x='Heart_Attack_Risk',
            y='Age',
            data=df)
plt.show()
```

```python

df['Age_Group'] = pd.cut(
df['Age'],
bins=[0,30,45,60,100],
labels=['Young',
'Adult',
'Middle Age',
'Senior']
)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Gender'] = le.fit_transform(df['Gender'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,18))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.countplot(x='Smoking',
              hue='Heart_Attack_Risk',
              data=df)
plt.show()
```

```python
sns.countplot(x='Diabetes',
              hue='Heart_Attack_Risk',
              data=df)
plt.show()
```
